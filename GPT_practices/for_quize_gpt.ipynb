```python
# function calling 요청 처리 흐름 정리 
# 1.질문 입력:  
# "How is the weather in Seoul?"라는 질문을 입력합니다.
# 이 질문은 chain.invoke({"city": "Seoul"})를 통해 전달됩니다.
#
# 2. 프롬프트 처리:
# 프롬프트 템플릿과 입력 데이터가 결합되어 모델에 전달될 프롬프트가 생성됩니다. 여기서는 "How is the weather in Seoul?"입니다.
#
# 3. 모델의 함수 호출 판단:
# ChatOpenAI 모델은 이 프롬프트를 분석하여 어떤 작업이 필요한지 결정합니다.
# 모델은 질문이 "Seoul"의 날씨를 묻는 것이므로, 이를 처리하기 위해 get_weather 함수를 호출해야 한다고 판단합니다.
#
# 4. 함수 호출 요청 생성:
# 모델은 함수 호출을 위해 필요한 매개변수를 포함한 요청을 생성합니다. 예를 들어, 위도와 경도를 포함하는 형태로:
# {
#    "function_call": {
#        "name": "get_weather",
#        "arguments": {
#            "lon": "126.9779692",
#            "lat": "37.566535"
#        }
#    }
# }
#
# 5. 함수 호출 정보 추출:
# 모델의 응답에서 함수 호출 정보 (function_call 객체)를 추출합니다.
# function_call 객체에서 호출할 함수 이름과 필요한 매개변수를 확인합니다.
#
#
# 6. 실제 함수 호출: 
# 추출한 매개변수를 사용하여 실제 get_weather 함수를 호출합니다.
# result = get_weather(lon="126.9779692", lat="37.566535")
# 
# 7. API 호출 및 결과 반환:
# get_weather 함수는 날씨 API를 호출하여 결과를 받아옵니다
#
# 8. 결과 출력:
# get_weather 함수가 반환한 결과를 출력합니다. 예를 들어:
# {
#    "weather": "Sunny",
#    "temperature": 25
# }

from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate

def get_weather(lon, tat):
    print("call an api ...")
 
function = {
    "name": "get_weather",
    "description": "function that takes longitude and latitude to find the weather of a place ", 
    "parameters": {
        "type": "object", 
        "properties": {
            "lon": { 
                "type": "string", 
                "description": "The longitude coordinate"
            }, 
            "lat": {
                "type": "string", 
                "description": "The latitude coordinate"
            }
        }
    },
    "required": ["lon", "lat"]
}

llm = ChatOpenAI(
    temperature=0.1
).bind(
    function_call="auto",
    functions=[
        function
    ]
)

prompt = PromptTemplate.from_template("Who is the weather in {city}")

chain = prompt | llm 

response = chain.invoke({
    "city": "Seoul"
})

response = response.additional_kwargs["function_call"]["arguments"]

import json 

r = json.loads(response)

get_weather(r['lon'], r['lat'])
```
